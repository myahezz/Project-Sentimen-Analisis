# Scraping Data dari Google Play

```python
# Melakukan Scraping 
link_aplication = 'jp.konami.pesam'

scraping = reviews (
    link_aplication,
    lang = 'id',
    country = 'id',
    sort = Sort.MOST_RELEVANT,
    count= 15000
)
# Menyimpan data hasil scraping
scraping_df = pd.DataFrame(scraping[0])
scraping_df.to_csv("ulasan_efootball2025_full.csv",index=False)
scraping_df.head()
```
